```python
def discounted_return(rewards, gamma):
    """
    Calcula a utilidade descontada:

        U = r0 + gamma*r1 + gamma^2*r2 + ... + gamma^t*rt

    Retorna:
    - utilidade total descontada
    - lista com a contribuição de cada recompensa
    """
    terms = []

    for t, r in enumerate(rewards):
        term = (gamma ** t) * r
        terms.append(term)

    return sum(terms), terms


# ============================================================
# Pergunta 3
# ============================================================

print("Pergunta 3")
print("-" * 50)

rewards_q3 = [-0.04, -0.04, -0.04, -0.04, 1.0]
gamma_q3 = 0.9

undiscounted_q3 = sum(rewards_q3)
discounted_q3, terms_q3 = discounted_return(rewards_q3, gamma_q3)

print("Recompensas:", rewards_q3)
print(f"Gamma: {gamma_q3}")
print()

print("Contribuições descontadas:")
for t, (r, term) in enumerate(zip(rewards_q3, terms_q3)):
    print(f"t={t}: gamma^{t} * r_{t} = {gamma_q3}^{t} * {r} = {term:.5f}")

print()
print(f"Recompensa total não descontada = {undiscounted_q3:.5f}")
print(f"Recompensa total descontada     = {discounted_q3:.5f}")


# ============================================================
# Pergunta 4
# ============================================================

print("\n\nPergunta 4")
print("-" * 50)

rewards_q4 = [0, 0, 0, 0, 1]

for gamma in [0.9, 0.5]:
    discounted_q4, terms_q4 = discounted_return(rewards_q4, gamma)

    print(f"Gamma: {gamma}")
    print("Contribuições descontadas:")

    for t, (r, term) in enumerate(zip(rewards_q4, terms_q4)):
        print(f"t={t}: gamma^{t} * r_{t} = {gamma}^{t} * {r} = {term:.5f}")

    print(f"Utilidade descontada = {discounted_q4:.5f}")
    print()
```

# Estatégia de exploração softmax

```python
import math

def softmax_boltzmann(q_values, tau):
    """
    Calcula as probabilidades softmax/Boltzmann:

        P(a | s) = exp(Q(s,a) / tau) / sum_a' exp(Q(s,a') / tau)

    Parâmetros:
    - q_values: dicionário {ação: valor_Q}
    - tau: temperatura (> 0)

    Retorna:
    - dicionário {ação: probabilidade}
    """
    if tau <= 0:
        raise ValueError("tau deve ser maior que zero.")

    numeradores = {}

    for action, q in q_values.items():
        numeradores[action] = math.exp(q / tau)

    denominador = sum(numeradores.values())

    probabilities = {}

    for action, num in numeradores.items():
        probabilities[action] = num / denominador

    return probabilities


# Exemplo: mesmo estado usado nas notas
q_values = {
    "Norte": 0.41,
    "Sul": -0.12,
    "Leste": 0.28,
    "Oeste": -0.04,
}

for tau in [1.0, 0.5, 0.1]:
    print(f"\nTemperatura tau = {tau}")
    print("-" * 40)

    probabilities = softmax_boltzmann(q_values, tau)

    for action, q in q_values.items():
        print(
            f"{action:5s}: "
            f"Q = {q: .2f}, "
            f"exp(Q/tau) = {math.exp(q / tau): .5f}, "
            f"P = {probabilities[action]:.4f}"
        )

    print(f"Soma das probabilidades = {sum(probabilities.values()):.4f}")
```
